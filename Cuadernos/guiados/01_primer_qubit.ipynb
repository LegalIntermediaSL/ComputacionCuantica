# El Primer Qubit

**Objetivo:** entender el estado de un qubit como vector en ℂ², las coordenadas de la esfera de Bloch y cómo funciona la medición.

**Sin hardware necesario** — todo con NumPy.

```python
import numpy as np

# Estado |0⟩ y |1⟩ en notación vectorial
zero = np.array([1, 0], dtype=complex)
one  = np.array([0, 1], dtype=complex)

print('|0⟩ =', zero)
print('|1⟩ =', one)
print('⟨0|0⟩ =', np.dot(zero.conj(), zero).real)  # norma = 1
```

## Superposición

Un qubit puede estar en cualquier combinación lineal α|0⟩ + β|1⟩ con |α|² + |β|² = 1.
El significado físico: |α|² es la probabilidad de medir 0 y |β|² la de medir 1.

```python
alpha = 1/np.sqrt(2)
beta  = 1/np.sqrt(2)
psi = alpha * zero + beta * one
print('|ψ⟩ =', psi)
print('Norma:', np.linalg.norm(psi))
print('P(0) =', abs(alpha)**2, '  P(1) =', abs(beta)**2)
```

## Esfera de Bloch

Todo estado puro se puede escribir como |ψ⟩ = cos(θ/2)|0⟩ + e^(iφ)sin(θ/2)|1⟩.
Los parámetros θ ∈ [0,π] y φ ∈ [0,2π) parametrizan la esfera unitaria.

```python
def bloch_coords(psi):
    alpha, beta = psi[0], psi[1]
    theta = 2 * np.arccos(np.clip(abs(alpha), 0, 1))
    phi = np.angle(beta) - np.angle(alpha)
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    return x, y, z

for name, state in [('|0⟩', zero), ('|1⟩', one), ('|+⟩', psi)]:
    x,y,z = bloch_coords(state)
    print(f'{name}: (x={x:.3f}, y={y:.3f}, z={z:.3f})')
```

## Medición

Medir proyecta el qubit a |0⟩ o |1⟩ con probabilidades |α|² y |β|². Simulemos 1000 mediciones:

```python
def medir(psi, n=1000):
    p0 = abs(psi[0])**2
    resultados = np.random.choice([0, 1], size=n, p=[p0, 1-p0])
    return {0: (resultados==0).sum(), 1: (resultados==1).sum()}

conteos = medir(psi)
print(f'Conteos tras 1000 mediciones: {conteos}')
print(f'P(0) empírico = {conteos[0]/1000:.3f}  (teórico = {abs(psi[0])**2:.3f})')
```

## Ejercicio

2. ¿Cuáles son sus coordenadas en la esfera de Bloch?
3. ¿Cuántas mediciones necesitas para distinguir |+⟩ de |−⟩ en la base computacional?

```python
# Tu solución aquí
minus = (zero - one) / np.sqrt(2)
print('|−⟩ =', minus)
print('Norma:', np.linalg.norm(minus))
print('Bloch:', bloch_coords(minus))
```
