# Problema resuelto 06: Bell y correlaciones en Qiskit

## Problema

Usar un circuito de Bell para explicar por que Qiskit devuelve principalmente `00` y `11` y como debe interpretarse esa salida.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
Statevector.from_instruction(qc)
```

## Discusion

El estado ideal generado es

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).
$$

Eso implica que, al medir en la base computacional, solo deben aparecer `00` y `11` en el caso ideal. La aparicion dominante de esos dos resultados no es accidental: refleja la estructura entrelazada del estado.


## Conclusion

Qiskit no solo permite dibujar el circuito, sino conectar de forma directa la estructura del estado ideal con las correlaciones observadas en la salida medida.
