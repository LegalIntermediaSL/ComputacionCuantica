# Problema resuelto 10: Grover y amplificacion en Qiskit

## Problema

Interpretar una versión mínima del esquema de Grover como un proceso de redistribución de amplitudes para favorecer un resultado de medición.


La intuición central de Grover no es que el circuito “pruebe todo a la vez” de forma ingenua, sino que reorganiza amplitudes para hacer más probable la observación del estado marcado.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)
qc.h([0, 1])
qc.cz(0, 1)
qc.h([0, 1])
qc.x([0, 1])
qc.cz(0, 1)
qc.x([0, 1])
qc.h([0, 1])
qc.measure([0, 1], [0, 1])
qc.draw('text')
```

## Conclusion

Aunque esta versión es muy pequeña, ya deja ver el patrón conceptual de Grover: marcar, difundir y medir para aumentar la probabilidad de una respuesta deseada.
