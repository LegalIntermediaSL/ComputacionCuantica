# Problema resuelto 02: estados de Bell

## Problema

Construir el estado de Bell $|\Phi^+\rangle$ con un circuito en Qiskit y describir los resultados de medicion esperados.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
qc.draw('text')
```

La salida ideal del circuito corresponde al estado

$$
|\Phi^+\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle).
$$

Por tanto, al medir muchas veces se esperan principalmente los resultados `00` y `11`, cada uno con probabilidad cercana a $1/2$.


## Analisis conceptual

Este ejemplo es importante porque el estado final no puede factorizarse en un producto de dos estados independientes. La CNOT no solo propaga informacion local: crea una correlacion genuinamente cuantica cuando el qubit de control llega en superposicion.


```python
import numpy as np

phi_plus = np.array([1 / np.sqrt(2), 0, 0, 1 / np.sqrt(2)], dtype=complex)
np.vdot(phi_plus, phi_plus)
```

La normalizacion confirma que el vector es un estado fisico valido. Su importancia, sin embargo, no esta solo en la norma sino en la imposibilidad de descomponerlo como producto tensorial de dos qubits separados.


## Conclusion

Este ejemplo muestra como un circuito muy corto puede generar entrelazamiento. La combinacion de `H` y `CNOT` es una de las construcciones mas importantes de la computacion cuantica elemental.
