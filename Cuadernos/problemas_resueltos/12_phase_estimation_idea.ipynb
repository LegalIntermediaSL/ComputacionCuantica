# Problema resuelto 12: idea de phase estimation

## Problema

Explicar por qué phase estimation actúa como puente entre unitarias, autovectores, QFT e información medible en un registro auxiliar.


La idea esencial es transformar una fase, que inicialmente está codificada de forma multiplicativa en una evolución unitaria, en información leíble en bits mediante qubits auxiliares y una QFT inversa.


## Conclusion

Phase estimation es importante porque reorganiza muchas piezas del curso en una sola arquitectura: control, fase, Fourier cuántica y medición final con significado informacional.
