# Problema resuelto 08: puertas de fase e interferencia en Qiskit

## Problema

Justificar por qué una puerta `Z` puede ser relevante aunque no cambie directamente las probabilidades de una medición inmediata en la base computacional.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)
qc.h(0)
Statevector.from_instruction(qc)
```

## Discusion

La puerta `Z` cambia la fase relativa entre las componentes de $|+\rangle$. Al aplicar una segunda `H`, esa diferencia de fase se traduce en una diferencia observable de amplitudes en la base computacional. Ahí es donde aparece la interferencia.


## Conclusion

Una puerta de fase puede parecer invisible en una medición concreta, pero deja huella en la evolución posterior del circuito. Esa es una de las ideas más importantes que Qiskit ayuda a visualizar.
