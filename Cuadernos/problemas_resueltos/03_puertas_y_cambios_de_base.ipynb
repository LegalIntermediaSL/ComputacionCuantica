# Problema resuelto 03: puertas y cambios de base

## Problema

Mostrar que la puerta Hadamard cambia entre la base computacional y la base formada por $|+\rangle$ y $|-\rangle$.


```python
import numpy as np

H = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)
ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)
H @ ket_0, H @ ket_1
```

## Discusion

Los resultados son exactamente los estados $|+\rangle$ y $|-\rangle$. Esto muestra que `H` no es solo una puerta util para crear superposicion, sino tambien una herramienta de cambio de base.


```python
ket_plus = (ket_0 + ket_1) / np.sqrt(2)
ket_minus = (ket_0 - ket_1) / np.sqrt(2)
H @ ket_plus, H @ ket_minus
```

Aplicar `H` una segunda vez devuelve la base computacional. En ese sentido, Hadamard actua como una traduccion reversible entre dos descripciones igualmente validas del mismo qubit.
