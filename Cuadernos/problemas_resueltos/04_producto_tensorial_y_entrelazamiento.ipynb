# Problema resuelto 04: producto tensorial y entrelazamiento

## Problema

Explicar por que el estado $|\Phi^+\rangle = (|00\rangle + |11\rangle)/\sqrt{2}$ no puede escribirse como producto tensorial de dos estados de un qubit.


Supongamos que pudiera escribirse como

$$
(a|0\rangle + b|1\rangle) \otimes (c|0\rangle + d|1\rangle).
$$

Al expandir, obtenemos

$$
ac|00\rangle + ad|01\rangle + bc|10\rangle + bd|11\rangle.
$$


Para reproducir $|\Phi^+\rangle$, deberiamos tener coeficientes nulos en `01` y `10`, pero no nulos en `00` y `11`. Eso fuerza condiciones incompatibles para los productos $ad$ y $bc$ sin destruir a la vez $ac$ o $bd$.


```python
import numpy as np

phi_plus = np.array([1 / np.sqrt(2), 0, 0, 1 / np.sqrt(2)], dtype=complex)
phi_plus
```

## Conclusion

La imposibilidad de factorizar este estado es precisamente lo que justifica llamarlo entrelazado. No basta con que dos qubits esten correlacionados: hace falta que el estado global no pueda separarse en dos estados locales.
