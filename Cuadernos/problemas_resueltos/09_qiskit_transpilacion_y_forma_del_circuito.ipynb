# Problema resuelto 09: transpilacion y forma del circuito

## Problema

Entender por qué un circuito transpiled puede tener una forma distinta al circuito original y, sin embargo, representar la misma tarea computacional ideal.


```python
from qiskit import QuantumCircuit, transpile

qc = QuantumCircuit(1, 1)
qc.h(0)
qc.measure(0, 0)

tqc = transpile(qc, basis_gates=['rz', 'sx', 'x'])
tqc.draw('text')
```

## Discusion

La transpilación puede reescribir una puerta abstracta en términos de otras puertas soportadas por una arquitectura concreta. El circuito cambia de forma, pero no de significado ideal si la equivalencia unitaria se mantiene.
