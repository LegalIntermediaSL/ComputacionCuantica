# Problema resuelto 01: superposicion y medicion

## Problema

Analizar el estado $|+\rangle = (|0\rangle + |1\rangle)/\sqrt{2}$ y calcular sus probabilidades de medicion en la base computacional.


Las amplitudes del estado son

$$
\alpha = \frac{1}{\sqrt{2}},
\qquad
\beta = \frac{1}{\sqrt{2}}.
$$

Por tanto,

$$
|\alpha|^2 = \frac{1}{2},
\qquad
|\beta|^2 = \frac{1}{2}.
$$


```python
import numpy as np

alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)
prob_0 = abs(alpha)**2
prob_1 = abs(beta)**2
prob_0, prob_1
```

## Conclusion

El estado $|+\rangle$ produce `0` y `1` con la misma probabilidad. Este es uno de los ejemplos mas simples de superposicion equilibrada.
