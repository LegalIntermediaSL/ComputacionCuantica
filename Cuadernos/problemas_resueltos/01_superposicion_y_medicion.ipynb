# Problema resuelto 01: superposicion y medicion

## Problema

Analizar el estado $|+\rangle = (|0\rangle + |1\rangle)/\sqrt{2}$ y calcular sus probabilidades de medicion en la base computacional.


Las amplitudes del estado son

$$
\alpha = \frac{1}{\sqrt{2}},
\qquad
\beta = \frac{1}{\sqrt{2}}.
$$

Por tanto,

$$
|\alpha|^2 = \frac{1}{2},
\qquad
|\beta|^2 = \frac{1}{2}.
$$


```python
import numpy as np

alpha = 1 / np.sqrt(2)
beta = 1 / np.sqrt(2)
prob_0 = abs(alpha)**2
prob_1 = abs(beta)**2
prob_0, prob_1
```

## Discusion

Este resultado no significa que el qubit estuviera ocultamente en `0` la mitad del tiempo y en `1` la otra mitad. La superposicion contiene informacion de fase y coherencia que solo se hace visible al combinar este estado con otras operaciones, por ejemplo con una puerta Hadamard adicional.


```python
hadamard = (1 / np.sqrt(2)) * np.array([[1, 1], [1, -1]], dtype=complex)
ket_plus = np.array([1 / np.sqrt(2), 1 / np.sqrt(2)], dtype=complex)
hadamard @ ket_plus
```

Aplicar `H` de nuevo a $|+\rangle$ devuelve el estado $|0\rangle$. Esa observacion muestra que la superposicion no era una mera mezcla clasica.


## Conclusion

El estado $|+\rangle$ produce `0` y `1` con la misma probabilidad, pero su estructura cuantica es mas rica que una incertidumbre clasica. La diferencia aparece cuando dejamos que el estado interfiera consigo mismo mediante nuevas puertas.
