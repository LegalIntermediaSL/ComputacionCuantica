# Problema resuelto 11: Statevector frente a DensityMatrix

## Problema

Explicar por qué `DensityMatrix` generaliza a `Statevector` y en qué situaciones esa generalización resulta necesaria.


```python
from qiskit.quantum_info import Statevector, DensityMatrix

psi = Statevector([1/2**0.5, 1/2**0.5])
rho = DensityMatrix(psi)
psi, rho
```

## Discusion

`Statevector` funciona muy bien para estados puros ideales. `DensityMatrix` permite además describir estados mixtos, trazas parciales y escenarios con pérdida de coherencia o de acceso a información.
