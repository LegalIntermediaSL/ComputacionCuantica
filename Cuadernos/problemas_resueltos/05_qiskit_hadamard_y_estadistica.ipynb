# Problema resuelto 05: Hadamard y estadistica en Qiskit

## Problema

Relacionar la puerta Hadamard con la preparacion del estado $|+\rangle$ y explicar por que una medicion repetida produce una distribucion aproximadamente uniforme.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0)
Statevector.from_instruction(qc)
```

## Discusion

La salida ideal del circuito es el estado

$$
|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle).
$$

Las amplitudes tienen el mismo modulo, por lo que las probabilidades de obtener `0` y `1` al medir son ambas $1/2$.


Si ejecutamos este circuito con muchos `shots`, las cuentas no seran exactamente iguales en cada experimento, pero deberian acercarse a una proporcion uniforme conforme aumente el numero de repeticiones.


## Conclusion

Este ejemplo muestra muy bien la diferencia entre una descripcion de estado y una descripcion estadistica de resultados. Ambas son correctas, pero responden a preguntas distintas.
