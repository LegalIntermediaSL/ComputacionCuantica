# Problema resuelto 07: medicion en base X con Qiskit

## Problema

Explicar cómo medir un qubit en la base $\{|+\rangle, |-\rangle\}$ usando únicamente las herramientas básicas de Qiskit.


La idea consiste en cambiar primero de base. Como la puerta `H` conecta la base computacional con la base X, medir en base X equivale a aplicar `H` y medir luego de forma estándar.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)
qc.h(0)
qc.h(0)
qc.measure(0, 0)
qc.draw('text')
```

## Conclusion

Qiskit no necesita un comando especial para cada base de medición elemental. A menudo basta con aplicar antes una puerta que traduzca la base deseada a la base computacional.
