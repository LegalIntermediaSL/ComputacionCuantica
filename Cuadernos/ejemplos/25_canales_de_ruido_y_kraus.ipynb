# Canales de ruido y operadores de Kraus

Cuaderno conceptual para conectar canales cuanticos, estados mixtos y ruido efectivo.

## Representacion de Kraus

Un canal cuantico puede escribirse como

$$
\mathcal{E}(\rho) = \sum_k E_k \rho E_k^\dagger.
$$

```python
import numpy as np

p = 0.2
K0 = np.sqrt(1 - p) * np.eye(2, dtype=complex)
K1 = np.sqrt(p) * np.array([[1, 0], [0, -1]], dtype=complex)

rho_plus = 0.5 * np.array([[1, 1], [1, 1]], dtype=complex)
rho_out = K0 @ rho_plus @ K0.conj().T + K1 @ rho_plus @ K1.conj().T

print("rho inicial:\n", rho_plus)
print("rho final:\n", rho_out)
print("traza final:", np.trace(rho_out))
```
