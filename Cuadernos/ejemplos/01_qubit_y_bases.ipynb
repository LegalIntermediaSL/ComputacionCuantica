# Ejemplo 01: qubit y bases

## Objetivo

Introducir la representacion vectorial de un qubit y verificar la normalizacion de algunos estados simples.


Un qubit puro puede escribirse como

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle,
$$

con la condicion

$$
|\alpha|^2 + |\beta|^2 = 1.
$$


```python
import numpy as np

ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)
ket_plus = (ket_0 + ket_1) / np.sqrt(2)

np.vdot(ket_plus, ket_plus)
```

El producto interno anterior debe devolver `1`, confirmando que el estado esta normalizado.
