# Ejemplo 01: qubit y bases

## Objetivo

Introducir la representacion vectorial de un qubit, verificar normalizacion y comparar la base computacional con la base de Hadamard.


Un qubit puro puede escribirse como

$$
|\psi\rangle = \alpha |0\rangle + \beta |1\rangle,
$$

con la condicion

$$
|\alpha|^2 + |\beta|^2 = 1.
$$

La base computacional esta formada por $|0\rangle$ y $|1\rangle$, mientras que otra base muy util es la formada por $|+\rangle$ y $|-\rangle$.


```python
import numpy as np

ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)
ket_plus = (ket_0 + ket_1) / np.sqrt(2)
ket_minus = (ket_0 - ket_1) / np.sqrt(2)

ket_plus, ket_minus
```

```python
norm_plus = np.vdot(ket_plus, ket_plus)
norm_minus = np.vdot(ket_minus, ket_minus)
overlap = np.vdot(ket_plus, ket_minus)
norm_plus, norm_minus, overlap
```

Los estados $|+\rangle$ y $|-\rangle$ estan normalizados y son ortogonales. Eso confirma que tambien constituyen una base valida del espacio de un qubit.


## Comentario didactico

La posibilidad de cambiar de base sin salir del mismo espacio vectorial es una de las claves para entender medicion y puertas cuanticas. En computacion cuantica, muchas operaciones utiles consisten precisamente en pasar de una base a otra y volver.
