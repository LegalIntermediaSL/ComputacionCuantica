# Ejemplo 16: idea de Sampler y Estimator

## Objetivo

Distinguir conceptualmente entre una interfaz orientada a distribuciones de salida y otra orientada a valores esperados de observables.


`Sampler` y `Estimator` representan dos formas diferentes de preguntar por el resultado de una computación cuántica.

- `Sampler`: orientado a muestreo y distribuciones.
- `Estimator`: orientado a valores esperados de observables.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1)
qc.h(0)
qc.draw('text')
```

En un proyecto más avanzado, el mismo circuito podría consultarse con dos intenciones distintas: obtener una distribución de salidas o estimar el valor esperado de un observable adecuado.
