# Ejemplo 19: Estimator conceptual

## Objetivo

Reforzar la idea de que algunos problemas en Qiskit no buscan una distribución de salidas, sino un valor esperado asociado a un observable.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp

qc = QuantumCircuit(1)
qc.h(0)
observable = SparsePauliOp.from_list([('X', 1.0)])
qc.draw('text'), observable
```

La estructura circuito + observable prepara de forma muy natural la idea de `Estimator`, especialmente en algoritmos variacionales y problemas de energía.
