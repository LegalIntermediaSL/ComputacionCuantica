# Estimator con observables distintos

Cuaderno conceptual para remarcar que un mismo estado puede producir valores esperados muy distintos segun el observable elegido.

```python
import numpy as np

zero = np.array([1, 0], dtype=complex)
one = np.array([0, 1], dtype=complex)
plus = (zero + one) / np.sqrt(2)
X = np.array([[0, 1], [1, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)

def expval(psi, A):
    return np.real_if_close(np.vdot(psi, A @ psi))

for nombre, estado in [("|0>", zero), ("|1>", one), ("|+>", plus)]:
    print(nombre, "<Z>=", expval(estado, Z), "<X>=", expval(estado, X))
```
