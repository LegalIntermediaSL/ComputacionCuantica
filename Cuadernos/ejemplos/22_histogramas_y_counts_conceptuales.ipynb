# Ejemplo 22: histogramas y counts conceptuales

## Objetivo

Reforzar la idea de que una distribución de counts no es un detalle accesorio, sino la forma natural en que se leen muchos resultados de circuitos medidos.


```python
counts = {'00': 510, '11': 514}
total = sum(counts.values())
probabilities = {k: v / total for k, v in counts.items()}
probabilities
```

Esta traducción de counts a probabilidades empíricas es una de las operaciones interpretativas más habituales cuando se trabaja con Qiskit.
