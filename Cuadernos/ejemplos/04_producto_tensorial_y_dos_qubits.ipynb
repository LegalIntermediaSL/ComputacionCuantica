# Ejemplo 04: producto tensorial y dos qubits

## Objetivo

Mostrar como se construyen estados de dos qubits a partir del producto tensorial y por que el espacio crece de dimension 2 a dimension 4.


```python
import numpy as np

ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)

ket_01 = np.kron(ket_0, ket_1)
ket_10 = np.kron(ket_1, ket_0)
ket_01, ket_10
```

Los vectores obtenidos viven ya en un espacio de cuatro componentes. Esta es la razon algebraica por la que el numero de amplitudes crece exponencialmente con el numero de qubits.


```python
basis = {
    '00': np.kron(ket_0, ket_0),
    '01': np.kron(ket_0, ket_1),
    '10': np.kron(ket_1, ket_0),
    '11': np.kron(ket_1, ket_1),
}
basis
```

Esta base es el punto de partida para describir estados de Bell, puertas controladas y algoritmos con varios qubits.
