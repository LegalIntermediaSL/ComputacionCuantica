# Estimator y Hamiltonianos sencillos

Cuaderno conceptual para enlazar observables, Hamiltonianos sencillos y estimacion de valores esperados.

## Objetivo

Entender por que un observable como $H = Z$ o $H = Z_1 Z_2$ puede verse como una cantidad a estimar y no solo como una expresion formal.

```python
import numpy as np

zero = np.array([1, 0], dtype=complex)
one = np.array([0, 1], dtype=complex)
plus = (zero + one) / np.sqrt(2)
Z = np.array([[1, 0], [0, -1]], dtype=complex)

def expval(psi, A):
    return np.real_if_close(np.vdot(psi, A @ psi))

for name, psi in [("|0>", zero), ("|1>", one), ("|+>", plus)]:
    print(name, "energia esperada para H=Z:", expval(psi, Z))
```
