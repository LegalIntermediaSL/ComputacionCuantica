# Ejemplo 11: transpilacion basica en Qiskit

## Objetivo

Mostrar que el circuito abstracto que diseñamos puede transformarse antes de ejecutarse para adaptarlo a restricciones o conjuntos de puertas soportados.


```python
from qiskit import QuantumCircuit, transpile

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

tqc = transpile(qc, basis_gates=['rz', 'sx', 'x', 'cx'])
tqc.draw('text')
```

La transpilacion no cambia la tarea computacional ideal del circuito, pero sí puede cambiar su forma operativa. Esta capa es esencial cuando se quiere pasar del modelo abstracto a hardware real o a un conjunto de puertas concreto.
