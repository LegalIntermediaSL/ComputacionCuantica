# Ejemplo 23: ideal vs ruidoso conceptual

## Objetivo

Comparar a nivel conceptual una distribución ideal y una distribución degradada por ruido.


```python
ideal = {'0': 0.5, '1': 0.5}
ruidoso = {'0': 0.57, '1': 0.43}
ideal, ruidoso
```

La diferencia entre ambas distribuciones es una primera forma intuitiva de visualizar cómo el ruido puede deformar una predicción ideal sin cambiar la estructura lógica del circuito subyacente.
