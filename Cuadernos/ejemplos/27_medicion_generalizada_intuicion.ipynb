# Medicion generalizada: intuicion

Cuaderno conceptual para introducir POVM y la idea de medicion mas general que la proyectiva.

## Punto de partida

Una POVM usa operadores positivos $E_i$ que satisfacen

$$
\sum_i E_i = I.
$$

La probabilidad del resultado $i$ se escribe como $p_i = \mathrm{Tr}(\rho E_i)$.

```python
import numpy as np

rho0 = np.array([[1, 0], [0, 0]], dtype=complex)
E0 = np.array([[0.7, 0.0], [0.0, 0.2]], dtype=complex)
E1 = np.eye(2, dtype=complex) - E0

p0 = np.trace(rho0 @ E0)
p1 = np.trace(rho0 @ E1)

print("p0 =", p0)
print("p1 =", p1)
print("suma =", p0 + p1)
```
