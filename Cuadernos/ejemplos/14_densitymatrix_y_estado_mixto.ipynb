# Ejemplo 14: DensityMatrix y estado mixto

## Objetivo

Dar una primera imagen práctica de cómo representar un estado puro y una mezcla estadística mediante matrices de densidad.


```python
from qiskit.quantum_info import Statevector, DensityMatrix

psi = Statevector([1, 0])
rho_pure = DensityMatrix(psi)
rho_pure
```

Una matriz de densidad puede representar tanto un estado puro como un estado mixto. Eso la convierte en una herramienta mejor adaptada a ruido, subsistemas y mezcla estadística.
