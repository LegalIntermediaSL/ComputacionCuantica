# Ejemplo 10: circuitos parametrizados en Qiskit

## Objetivo

Introducir la idea de circuitos con parametros simbolicos, muy utiles en algoritmos variacionales y en exploracion de familias de estados.


```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter

theta = Parameter('theta')
qc = QuantumCircuit(1, 1)
qc.ry(theta, 0)
qc.measure(0, 0)
qc
```

Este tipo de circuito no fija de inmediato una rotacion concreta. En su lugar, define una familia continua de circuitos que pueden ligarse despues a valores numericos del parametro.


```python
qc_bound = qc.assign_parameters({theta: 1.234})
qc_bound.draw('text')
```
