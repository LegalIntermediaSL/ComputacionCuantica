# Ejemplo 09: puertas de fase en Qiskit

## Objetivo

Ilustrar que una puerta de fase puede ser físicamente relevante aunque no cambie inmediatamente las probabilidades en la base computacional.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0)
qc.z(0)
state = Statevector.from_instruction(qc)
state
```

La puerta `Z` transforma $|+\rangle$ en $|-\rangle$. En la base computacional ambos estados tienen las mismas probabilidades, pero su fase relativa es distinta y eso afecta la interferencia posterior.


```python
qc.draw('text')
```
