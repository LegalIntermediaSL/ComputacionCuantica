# Fidelidad antes y despues de ruido

Cuaderno conceptual para comparar un estado objetivo con una version degradada por ruido efectivo.

```python
import numpy as np

rho_ideal = 0.5 * np.array([[1, 1], [1, 1]], dtype=complex)
rho_ruidoso = 0.5 * np.array([[1, 0.6], [0.6, 1]], dtype=complex)

print("Estado ideal:\n", rho_ideal)
print("Estado ruidoso:\n", rho_ruidoso)
print("Objetivo: discutir como la perdida de coherencia afecta a la cercania con el objetivo.")
```
