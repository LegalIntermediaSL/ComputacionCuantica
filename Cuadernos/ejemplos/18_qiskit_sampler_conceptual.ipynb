# Ejemplo 18: Sampler conceptual

## Objetivo

Entender cuándo tiene sentido pensar un flujo de trabajo de Qiskit como un problema de muestreo de distribuciones de salida.


`Sampler` es natural cuando el resultado de interés son strings de bits y sus probabilidades o frecuencias relativas.

Esta situación aparece, por ejemplo, cuando el circuito termina en medición y queremos reconstruir la distribución de salida del experimento.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.draw('text')
```

En una capa moderna de Qiskit, una consulta tipo `Sampler` expresaría muy bien la intención de obtener la distribución asociada a este circuito.
