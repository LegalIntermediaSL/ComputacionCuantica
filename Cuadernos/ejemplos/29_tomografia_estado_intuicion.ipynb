## Idea central

Un estado general de un qubit no se reconstruye desde una sola base de medida. Hace falta combinar informacion obtenida en distintas bases.

```python
bases = ["Z", "X", "Y"]
for base in bases:
    print("Medir en base", base, "aporta una componente distinta de informacion sobre el estado.")
```
