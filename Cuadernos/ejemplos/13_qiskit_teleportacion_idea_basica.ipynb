# Ejemplo 13: idea basica de teleportacion en Qiskit

## Objetivo

Mostrar la estructura general de un circuito de teleportacion cuantica y su dependencia del entrelazamiento y de la comunicacion clasica.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 2)
qc.h(1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.h(0)
qc.measure(0, 0)
qc.measure(1, 1)
qc.draw('text')
```

Este esquema deja visible la arquitectura del protocolo: preparación de un par entrelazado, acoplamiento con el estado a transferir y medición que produce los bits clásicos necesarios para las correcciones finales.
