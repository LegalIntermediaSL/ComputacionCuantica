# Ejemplo 08: medicion en distintas bases con Qiskit

## Objetivo

Mostrar que medir un mismo estado en bases distintas produce informacion distinta y que en Qiskit el cambio de base puede implementarse con puertas sencillas antes de medir.


Medir en la base computacional no es lo mismo que medir en la base de Hadamard. Para medir en la base $\{|+\rangle, |-\rangle\}$ suele aplicarse una puerta `H` antes de la medicion estandar.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(1, 1)
qc.h(0)
qc.h(0)
qc.measure(0, 0)
qc.draw('text')
```

En este circuito, la primera `H` prepara $|+\rangle$ a partir de $|0\rangle$. La segunda `H` cambia nuevamente de base y devuelve el estado a $|0\rangle$ antes de medir.
