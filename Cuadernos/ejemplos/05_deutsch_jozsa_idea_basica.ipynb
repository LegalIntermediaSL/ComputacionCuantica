# Ejemplo 05: idea basica de Deutsch-Jozsa

## Objetivo

Introducir la intuicion del algoritmo de Deutsch-Jozsa como ejemplo temprano de interferencia cuantica usada para extraer informacion global sobre una funcion.


El interes principal de Deutsch-Jozsa es pedagogico. Muestra que un circuito cuantico puede reorganizar amplitudes de manera que una medicion final revele una propiedad global del problema sin inspeccionar casos uno por uno en el sentido clasico.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 1)
qc.h(0)
qc.x(1)
qc.h(1)
qc.cx(0, 1)
qc.h(0)
qc.measure(0, 0)
qc.draw('text')
```

Este circuito no pretende aun ser una presentacion completa del algoritmo general, pero si da una primera imagen de como una secuencia de Hadamards, una consulta tipo oraculo y una medicion final puede revelar estructura global de la funcion.
