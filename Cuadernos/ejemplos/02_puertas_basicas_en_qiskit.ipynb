# Ejemplo 02: puertas basicas en Qiskit

## Objetivo

Construir circuitos elementales con Qiskit y visualizar el efecto de las puertas `H`, `X` y `Z`.


```python
from qiskit import QuantumCircuit

qc_h = QuantumCircuit(1, 1)
qc_h.h(0)
qc_h.measure(0, 0)
qc_h
```

Este circuito prepara el estado $|+\rangle$ y luego lo mide en la base computacional. Si se ejecuta muchas veces, se esperan resultados cercanos a una distribucion uniforme entre `0` y `1`.


```python
qc_x = QuantumCircuit(1, 1)
qc_x.x(0)
qc_x.measure(0, 0)

qc_z = QuantumCircuit(1, 1)
qc_z.z(0)
qc_z.measure(0, 0)

qc_x.draw('text'), qc_z.draw('text')
```

La puerta `X` cambia directamente el bit cuantico en la base computacional, mientras que la puerta `Z` modifica una fase relativa. Esa diferencia es central: no todas las puertas se revelan de la misma manera en una sola medicion.
