# Ejemplo 06: Qiskit, estado vector y medicion

## Objetivo

Distinguir entre la descripcion ideal de un estado cuantico y los resultados estadisticos de una medicion en Qiskit.


Un mismo circuito puede estudiarse de dos maneras complementarias:

- mediante el estado ideal que produce antes de medir;
- mediante las cuentas obtenidas tras muchas mediciones.

Esta distincion es central para no confundir amplitudes con frecuencias observadas.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(1)
qc.h(0)

state = Statevector.from_instruction(qc)
state
```

El resultado ideal corresponde al estado $|+\rangle$. Si despues midiéramos este circuito muchas veces, obtendríamos aproximadamente una distribucion 50-50 entre `0` y `1`, pero eso ya es una capa estadistica distinta.
