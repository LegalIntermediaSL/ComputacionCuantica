# Ejemplo 21: SparsePauliOp basico

## Objetivo

Introducir de forma breve una representacion util de observables y Hamiltonianos sencillos en Qiskit.


```python
from qiskit.quantum_info import SparsePauliOp

hamiltonian = SparsePauliOp.from_list([('Z', 1.0), ('X', 0.25)])
hamiltonian
```

Esta capa resulta especialmente importante cuando el proyecto se mueve hacia estimacion de observables, VQE y formulaciones mas cercanas a Hamiltonianos fisicos.
