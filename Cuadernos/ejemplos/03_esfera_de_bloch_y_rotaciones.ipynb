# Ejemplo 03: esfera de Bloch y rotaciones

## Objetivo

Relacionar la parametrizacion de un qubit puro con la esfera de Bloch y observar como cambian sus componentes al variar los angulos $\theta$ y $\phi$.


Todo estado puro de un qubit puede escribirse como

$$
|\psi\rangle = \cos\left(\frac{\theta}{2}\right)|0\rangle + e^{i\phi}\sin\left(\frac{\theta}{2}\right)|1\rangle.
$$


```python
import numpy as np

theta = np.pi / 3
phi = np.pi / 4
state = np.array([
    np.cos(theta / 2),
    np.exp(1j * phi) * np.sin(theta / 2)
], dtype=complex)
state
```

```python
np.vdot(state, state)
```

La normalizacion debe ser uno. Esta forma parametrica es util porque permite visualizar el estado como un punto de la esfera de Bloch y entender las puertas como rotaciones geométricas.
