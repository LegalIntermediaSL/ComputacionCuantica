# Ejemplo 12: Grover minimo en Qiskit

## Objetivo

Dar una primera intuicion de la amplificacion de amplitud en una version muy pequeña del esquema de Grover.


Este cuaderno no pretende construir una implementación completa y general de Grover, sino mostrar cómo aparece la idea de marcar un estado y reorganizar amplitudes para favorecer su observación.


```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)
qc.h([0, 1])
qc.cz(0, 1)
qc.h([0, 1])
qc.x([0, 1])
qc.cz(0, 1)
qc.x([0, 1])
qc.h([0, 1])
qc.measure([0, 1], [0, 1])
qc.draw('text')
```
