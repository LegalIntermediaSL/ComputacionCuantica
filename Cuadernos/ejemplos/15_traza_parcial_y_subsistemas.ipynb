# Ejemplo 15: traza parcial y subsistemas

## Objetivo

Mostrar cómo un estado global entrelazado puede producir estados reducidos mixtos al tomar subsistemas.


```python
from qiskit.quantum_info import DensityMatrix, partial_trace
import numpy as np

phi_plus = np.array([1/np.sqrt(2), 0, 0, 1/np.sqrt(2)], dtype=complex)
rho = DensityMatrix(phi_plus)
partial_trace(rho, [1])
```

Este ejemplo resume una idea central de la información cuántica: el estado global puede ser puro mientras que el subsistema aislado aparece mixto.
