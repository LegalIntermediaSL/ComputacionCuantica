# DensityMatrix y open quantum systems

Cuaderno conceptual para conectar descripcion por matrices de densidad con sistemas abiertos y perdida de coherencia.

```python
import numpy as np

rho_puro = 0.5 * np.array([[1, 1], [1, 1]], dtype=complex)
rho_decoherente = 0.5 * np.array([[1, 0.3], [0.3, 1]], dtype=complex)

print("Estado ideal:\n", rho_puro)
print("Estado con menor coherencia:\n", rho_decoherente)
print("Objetivo: discutir por que esta descripcion ya no cabe comodamente en un unico statevector puro.")
```
