# Ejemplo 17: observables y Estimator

## Objetivo

Reforzar la idea de que algunos flujos de Qiskit no piden strings de salida, sino valores esperados de observables definidos sobre el estado preparado.


```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp

qc = QuantumCircuit(1)
qc.h(0)
obs = SparsePauliOp.from_list([('X', 1.0)])
qc.draw('text'), obs
```

La pareja circuito-observable es la estructura natural de trabajo cuando queremos pensar en `Estimator`.
