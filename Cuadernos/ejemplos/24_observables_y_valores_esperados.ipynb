# Observables y valores esperados

Este cuaderno conecta observables, valores esperados y Hamiltonianos sencillos.

## Idea central

Para un observable $A$ y un estado $|\psi\rangle$, el valor esperado es

$$
\langle A \rangle = \langle \psi | A | \psi \rangle.
$$

```python
import numpy as np

zero = np.array([1, 0], dtype=complex)
one = np.array([0, 1], dtype=complex)
plus = (zero + one) / np.sqrt(2)

X = np.array([[0, 1], [1, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)

def valor_esperado(psi, A):
    return np.vdot(psi, A @ psi)

for nombre, estado in [("|0>", zero), ("|1>", one), ("|+>", plus)]:
    print(nombre, "<Z>=", valor_esperado(estado, Z), "<X>=", valor_esperado(estado, X))
```
