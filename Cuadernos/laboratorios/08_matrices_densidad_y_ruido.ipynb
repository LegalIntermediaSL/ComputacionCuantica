# Matrices de Densidad, Traza Parcial y Entropía en Qiskit

En esta práctica exploraremos computacionalmente los conceptos vistos en la teoría de *Información Cuántica*. Usaremos el módulo `qiskit.quantum_info` para interactuar con matrices de densidad explícitas, trazas parciales y entropía termodinámica.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, DensityMatrix, partial_trace, entropy
import numpy as np
```

## 1. De Vectores de Estado a Matrices de Densidad
Vamos a definir el estado $|+\rangle$ y visualizar su matriz de densidad pura.

```python
qc_plus = QuantumCircuit(1)
qc_plus.h(0)

rho_plus = DensityMatrix(qc_plus)
print("Matriz de Densidad r_plus (Estado Puro):")
display(rho_plus.draw('latex'))
```

Comparemos esto con un estado máximamente mezclado (caos total). 
Lo construimos matemáticamente mezclando probabilísticamente protectores a base al $50\%$.

```python
rho_mixed = 0.5 * DensityMatrix.from_label('0') + 0.5 * DensityMatrix.from_label('1')
print("Matriz de Densidad r_mixed (Estado Mezclado máximo):")
display(rho_mixed.draw('latex'))

print(f"Pureza de r_plus: {rho_plus.purity():.2f} (Es exactamente 1)")
print(f"Pureza de r_mixed: {rho_mixed.purity():.2f} (Es menos de 1, concretamente 1/d)")
```

## 2. Traza Parcial sobre los Estados de Bell
Vamos a generar un par entrelazado de Bell $|\Phi^+\rangle_{AB}$ y a extraer el conocimiento local del qubit A mediante la técnica de la Traza Parcial.

```python
qc_bell = QuantumCircuit(2)
qc_bell.h(0)
qc_bell.cx(0, 1)

rho_global = DensityMatrix(qc_bell)
print("Matriz de densidad Global AB:")
display(rho_global.draw('latex'))

# Ejecutamos la traza parcial quitando uno de los qubits 
# Qiskit numera los qubits como 0 y 1. Borramos el subsistema de índice 1.
rho_A = partial_trace(rho_global, [1])

print("Matriz reducida del subsistema A:")
display(rho_A.draw('latex'))
```

## 3. Entropía de von Neumann
Comparamos la entropía global con la entropía residual una vez que el observador pierde datos globales.

```python
print(f"Entropía Global S(rho_AB) = {entropy(rho_global):.4f}")
print(f"Entropía del Subsistema S(rho_A) = {entropy(rho_A):.4f}")

print("\nLa entropía del subsistema es no-nula, confirmando el entrelazamiento termodinámico original a pesar de tratarse de un universo originario 'puro'.")
```
