# Laboratorio 11: QAOA y paisaje de coste

## Objetivo

Explorar de manera cualitativa la idea de paisaje de coste en QAOA y por que la eleccion de parametros importa tanto.


```python
import numpy as np

gammas = np.linspace(0, np.pi, 5)
betas = np.linspace(0, np.pi/2, 5)
landscape = [[float(np.cos(g) * np.sin(2*b)) for b in betas] for g in gammas]
landscape
```

## Discusion

El objetivo aqui no es modelar un Hamiltoniano realista completo, sino entrenar intuicion sobre la dependencia conjunta en varios parametros y sobre la necesidad de un optimizador clasico bien planteado.
