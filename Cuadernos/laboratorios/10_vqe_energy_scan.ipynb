# Laboratorio 10: VQE y barrido de energia

## Objetivo

Usar un circuito parametrizado sencillo para estudiar como cambia el valor esperado de un Hamiltoniano elemental al variar un parametro.


```python
import numpy as np

thetas = np.linspace(0, 2 * np.pi, 9)
energies = np.cos(thetas)
list(zip(thetas.round(3), energies.round(3)))
```

## Discusion

En un VQE muy simplificado, el objetivo del optimizador clasico es localizar regiones de parametros donde la energia esperada disminuye. Este barrido manual permite ver la geometria del problema antes de introducir un optimizador automatizado.
