# Laboratorio 09: QAOA guiado

## Objetivo

Visualizar la estructura de capas de QAOA y su relación con un Hamiltoniano de coste y un mezclador.


```python
from qiskit import QuantumCircuit
from qiskit.circuit import Parameter

gamma = Parameter('gamma')
beta = Parameter('beta')

qc = QuantumCircuit(2)
qc.h([0, 1])
qc.rzz(2 * gamma, 0, 1)
qc.rx(2 * beta, 0)
qc.rx(2 * beta, 1)
qc.draw('text')
```

## Preguntas de trabajo

1. Qué representa la capa de coste?
2. Qué representa el mezclador?
3. Por qué QAOA se presta naturalmente a una optimización híbrida?
