# Laboratorio guiado: trotterizacion y evolucion

Este laboratorio introduce la idea de aproximar una evolucion compleja mediante bloques mas simples.

## Idea de partida

Si $H = A + B$, entonces en general

$$
e^{-i(A+B)t} \neq e^{-iAt} e^{-iBt}.
$$

La trotterizacion usa productos de pasos pequenos para aproximar la evolucion total.

```python
import numpy as np

X = np.array([[0, 1], [1, 0]], dtype=complex)
Z = np.array([[1, 0], [0, -1]], dtype=complex)
t = 0.5
n = 4

print("Hamiltoniano esquematico H = X + Z")
print("Tiempo:", t)
print("Pasos de Trotter:", n)
print("Objetivo del laboratorio: comparar la intuicion de un solo paso frente a varios pasos pequeños.")
```
